## Import the needed package

```python
import sys
sys.path.insert(0, "./../")

import tensorflow_datasets as tfds
from utilities import count_labels, show_examples
```

## Extract the data

```python
(train_raw, test_raw), info = tfds.load(
    "cifar10",
    split=['train', 'test'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True
)
```

```python
trainset = train_raw.take(10000)
valset = test_raw.take(5000)
testset = test_raw.skip(5000)
```

## Explore the dataset

```python
info
```

```python
# class name
class_names = info.features['label'].names
print("Class Name:",
        ", ".join(class_names),
        f"Class Count: {len(class_names)}",
        sep="\n", end="\n\n")

# dataset size
print("Dataset Size:",
        f"Training data: {len(trainset)}",
        f"Validation data: {len(valset)}",
        f"Testing data: {len(testset)}",
        sep="\n", end="\n\n")

# count labels data count
print("Label Data Count:",
        f"trainset: {count_labels(trainset)}",
        f"valset: {count_labels(valset)}",
        f"testset: {count_labels(testset)}",
        sep="\n", end="\n\n")
```

```python
print("Trainset Sample")
show_examples(trainset, class_names, 3)
```

```python
print("Validationset Sample")
show_examples(valset, class_names, 3)
```

```python
print("Testset Sample")
show_examples(testset, class_names, 3)
```
